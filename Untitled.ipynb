```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_graph(x1, x2, x3, x4, x5, y1, y2, y3, y4, y5, xlabel, ylabel, title= "default"):
    
    plt.plot(x1, y1, label = "heuristic 1")
    plt.plot(x2, y2, label = "heuristic 2")
    plt.plot(x3, y3, label = "heuristic 3")
    plt.plot(x4, y4, label = "heuristic 4")
    plt.plot(x5, y5, label = "heuristic 5")
 
    # naming the x axis
    plt.xlabel(xlabel)
    
    # naming the y axis
    plt.ylabel(ylabel)
 
    # giving a title to the graph
    plt.title(title)
 
    plt.legend()
    
    # displaying the plot
    plt.show()
```

```python
def normalize_arr(arr):
    arr_np = np.array(arr)

    # Find the minimum and maximum values in the array
    min_val = np.min(arr_np)
    max_val = np.max(arr_np)
    if min_val == max_val:
        return arr
    # Normalize the array values between 0 and 1
    normalized_arr = (arr_np - min_val) / (max_val - min_val)
    return normalized_arr
```
